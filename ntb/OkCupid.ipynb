<span style="color: #57A6DC; font-weight: bold;">Ethnicity</span> is a
categorical feature which can take many different combinations of ehtnicities. It appears people have been allowed to self-identify with any
and all ethnicities (from a number of categories). Unfortunately, to have all
these ethnicities reported in one feature is not tidy and does not support
effective analysis. The self-identified membership of ethnicity is extracted to
different boolean features for each individual <span style="color: #57A6DC;
font-weight: bold;">Ethnicity</span>.

````python
# Define column of interest.
COLUMN = 'Ethnicity'

# Identify all unique combinations of ethnicities.
ethnicities = [e.split(', ') for e in df[COLUMN].dropna().unique()]

# Identify all individual unique ethnicities.
ethnicities = sorted(set([e for sublist in ethnicities for e in sublist]))

# Identify all unique ethnicities.
ethnicities = [e.strip().capitalize() for e in ethnicities]

# Define a function to create dictionaries of ethnicity.
def parse_ethnicity(x):
    if pd.isna(x):
        pass
    else:
        x = [x.strip().capitalize() for x in x.split(',')]
        return {e:True if e in x else False for e in ethnicities}

# Redefine 'Ethnicity' as a string data type for parsing.
df[COLUMN] = df[COLUMN].astype('str')

# Convert 'Ethnicity' values to dictionaries.
df[COLUMN] = df[COLUMN].apply(parse_ethnicity)

# Expand 'Ethnicity' into separate columns and join them to the DataFrame.
df = df.join(df[COLUMN].apply(pd.Series))

# Define a list of ethnicity columns.
eth_columns = [f'{COLUMN}: {e}' for e in ethnicities]

# Rename the ethnicity columns.
df = df.rename(columns={e: c for e, c in zip(ethnicities, eth_columns)})

# Redefine the ethnicity columns as categorical.
df[eth_columns] = df[eth_columns].astype('category')

# Redefine the ethnicity columns as a categorical.
plt_df[eth_columns] = df[eth_columns].astype('category')

# Redefine the ethnicity columns as a categorical.
mle_df[eth_columns] = df[eth_columns].astype('category')
````

indicate that it is actually a binned variable, rather than a continuous
variable. Therefore <span style="color: #57A6DC; font-weight:
bold;">Income</span> is converted to a categorical data type. Furthermore the
last three values of `250,000`, `500,000`, and `1,000,000` are very sparse and
are simply replaced with `250,000`. This means <span style="color: #57A6DC;
font-weight: bold;">Income</span> can take one of ten orderd values: `20,000`,
`30,000`, `40,000`, `50,000`, `60,000`, `70,000`, `80,000`, `100,000`,

````python
# Define column of interest.
COLUMN = 'Language skills'

# Define a function to create dictionaries of language proficiency pairs.   
def parse_proficiency(x):
    if x == 'nan':
        return {}
    
    # Dictionary to store skill and proficiency
    lp_dict = {}
    lps = x.split(',')

    for lp in lps:

        # Check if the entry has a proficiency level in parentheses
        if '(' in lp:
        
            # Split each entry into skill name and proficiency
            language, proficiency = lp.split('(')
            # proficiency = proficiency.strip('() ')
        else:
            # Default proficiency if none specified
            language = lp
            proficiency = 'unqualified'
            
        proficiency = proficiency.strip('() ')
        language = language.strip().capitalize()
        
        # Add to the dictionary
        lp_dict[language] = proficiency
    
    return lp_dict

# Redefine 'Languages' as a string data type for parsing.
df[COLUMN] = df[COLUMN].astype('str')

# Convert 'Languages' values to dictionaries.
df[COLUMN] = df[COLUMN].map(parse_proficiency)

# Expand 'Languages' into separate columns and join them to the DataFrame.
df = df.join(df[COLUMN].apply(pd.Series))

# Identify all unique languages.
languages = sorted(set(l for lp in df[COLUMN] for l in lp.keys()))

# Define a list of new names for ethnicity columns.
lpr_columns = [f'Language: {l}' for l in languages]

# Rename the language columns.
df = df.rename(columns={l: c for l, c in zip(languages, lpr_columns)})

# Reorder the columns alphabetically
df = df[sorted(df.columns)]

# Redefine the language columns as categoricals.
mle_df[lpr_columns] = df[lpr_columns].astype('category')

# Redefine the language columns as categoricals.
plt_df[lpr_columns] = df[lpr_columns].astype('category')
````

In order to test an association between a continuous variable and a non-dichotomous  categorical variable, 
one could use multiple two-sample t-tests. 
However, running multiple tests increases the probability of a false positive (type I error) 
so that it is greater than the significance threshold for each test. 
To avoid this issue, a better solution is to run an ANOVA; 
then, if the p-value for the ANOVA is significant, run Tukey’s range test.

(1) All observations should be independently and randomly sampled
(2) The standard deviations of the groups should be equal
(3) The data should be normally distributed or the sample size should be large
(4) The groups created by the categorical variable should be independent

Lambda is a Proportional Reduction in Error (PRE) measure which is interpreted as 
the amount of variance accounted for in predicting the dependent variable that can be attributed by the independent variable. 
Another way to say this is: 
how much better is our guess about which category of the dependent variable each case will fall into if 
we know the case’s value on the independent variable? 
Much like R-squared in regression (which is also a PRE measure), lambda is often represented as a percentage. 
Lambda is a directional measure in that the calculation differs based on which variable is treated as the independent variable.

**Note**: 
Higher kurtosis corresponds to greater extremity of deviations (or outliers). The kurtosis of any univariate normal distribution is 3. It is common to compare the kurtosis of a distribution to this value. Distributions with kurtosis less than 3 are said to be platykurtic, although this does not imply the distribution is "flat-topped" as is sometimes stated. Rather, it means the distribution produces fewer and less extreme outliers than does the normal distribution. An example of a platykurtic distribution is the uniform distribution, which does not produce outliers. Distributions with kurtosis greater than 3 are said to be leptokurtic. An example of a leptokurtic distribution is the Laplace distribution, which has tails that asymptotically approach zero more slowly than a Gaussian, and therefore produces more outliers than the normal distribution. It is also common practice to use an adjusted version of Pearson's kurtosis, 
the excess kurtosis, which is the kurtosis minus 3, to provide the comparison to the standard normal distribution. Source: [Wikipedia](https://en.wikipedia.org/wiki/Kurtosis)

### Dimensionality reduction
Also called Feature extraction. There are two main issues with high dimensional data: data sparsity and distance concentration. Dimensionaltity reduction aims to mitigate the consequences of these issues by combining the existing feature space into a new subset of features that are ordered by decreasing importance. This smaller subset saves a lot of computational time and resources while maintaining or even improving performance. The lack of interpretability is one of the drawbacks of dimensionality reduction.

## Data sparsity
A high-dimensional dataset will require a huge sample in order to represent all possible combinations of features. Because a supervised model can only learn from what is contained in the training dataset, this can lead to overfitting and lack of generalizability.

## Distance concentration
Many machine learning algorithms, such as k-nearest-neighbors and unsupervised clustering methods, involve calculating distances between observations. As the number of dimensions increases, the distance between all pairs of data points becomes increasingly similar. In other words, it becomes more difficult to discern whether two observations are any more or less similar to each other than to any other observation.

## Principal Component Analysis (PCA)
Principal component analysis (PCA) is an unsupervised learning algorithm that transforms high-dimensional data into a smaller number of features using principal components (PCs). The goal is to preserve as much variance and information (95%) from the original data in a lower-dimensional space. The new features generated by PCA are linear combinations of the original features. As such PCA, finds the subspace that is best for linear regression problems.

## Linear Discriminant Analysis (LDA)
Linear discriminant analysis (LDA) is a dimensionality reduction technique used for classification problems with continuous independent variables. LDA can be used to transform data into new features that are linear combinations of the original ones. However, instead of maximizing the preserved variance from the original data as in PCA, LDA works by maximizing the the ratio of variance between the classes to variance within the classes, such that they are well separated in lower dimensional space.
Assumptions: independent variables are multivariate normal and not multicolinear, dependent variables are homoscedastic, observations are statistically independent, sensitive to outliers, prone to overfitting (use shrinkage regularization). 

## Quadratic Discriminant Analysis (QDA)
Quadratic Discriminant Analysis (QDA) is very similar to LDA, but it can be used when classes have different covariance matrices. It does not require the assumption of homoscedasticity for the classes (similar distribution).

## T-distributed stochastic neighbor embedding (t-SNE)
t-SNE is another dimensionality reduction technique, but it uses a nonlinear method to map each data point from a high-dimensional space to a 2- or 3-dimensional space. As its name implies, t-SNE uses the student t-distribution to model the points in a way such that similar points are mapped closer to each other and dissimilar points are further apart. As a result, t-SNE is great at preserving the local structure of the original data.
